```python
import numpy as np # math
import pandas as pd # files
import matplotlib.pyplot as plt # graph
```

```python
path = r'D:\smoke_detection_iot.csv'
data = pd.read_csv(path)
```

```python
counts = data.groupby('Fire_Alarm').size()

mylabels = ["yes_Fire", "No_fire"]
```

```python
counts = data.groupby('Fire_Alarm').size()

mylabels = ["yes_Fire", "No_fire"]

# Create pie chart.
fig, ax = plt.subplots()

# Plot pie chart.
plt.pie(counts,colors=["orange", "mediumpurple"], labels = mylabels, autopct='%1.1f%%',startangle = 90, 
        wedgeprops = {"edgecolor" : "black",'linewidth': 0.5,'antialiased': True})


# Set autopct text color.
for text in ax.texts:
    text.set_color('white')

# Show plot.
plt.legend(frameon = 0,loc='upper right')
plt.title("Fire alarm graph", loc='left',color='gray')
plt.show()
```
